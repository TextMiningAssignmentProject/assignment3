```python
###http://www.e-jamet.org/_common/do.php?a=full&b=52&bidx=292&aidx=3578
from sklearn.feature_extraction.text import CountVectorizer
from math import log
import lib.getURL as url
from konlpy.tag import Okt  # 명사 추출기
import pandas as pd
okt = Okt()

getLocation = "./data/json"
saveLocation = "./data"
```

```python
def get_noun(doc_list):
    noun_list =[]
    for i in range(len(doc_list)):
        noun_list.append(okt.nouns(doc_list[i]))
    return noun_list
```

```python
# 카테고리별 텍스트 값 가져오기
booksText = url.getCatText(saveLocation, "books.txt")
selfDev = url.getCatText(saveLocation, "selfdev.txt")
program = url.getCatText(saveLocation, "programming.txt")
volunteer = url.getCatText(saveLocation, "volunteer.txt")

#카테고리 문서별 리스트
books_doc = booksText.split('|')
self_doc = selfDev.split('|')
program_doc = program.split('|')
volunteer_doc = volunteer.split('|')

#카테고리 문서별 명사 추출
books_noun = get_noun(books_doc)
self_noun = get_noun(self_doc)
program_noun = get_noun(program_doc)
volunteer_noun = get_noun(volunteer_doc)

# 카테고리별 문서 수
books_count = len(books_doc)
self_count = len(self_doc)
program_count = len(program_doc)
volunteer_count = len(volunteer_doc)

# 카테고리별 명사 추출
booksText = okt.nouns(booksText)
selfDev = okt.nouns(selfDev)
program = okt.nouns(program)
volunteer = okt.nouns(volunteer)
```

```python
#전체 문서
merged_corpus = booksText + selfDev + program + volunteer
merged_doc = books_doc + self_doc + program_doc + volunteer_doc
print(merged_corpus)
corpus_anoun = list(set(merged_corpus))

# 2개 이상 명사만 추출
corpus_result = []
for c in corpus_anoun:
    if (len(c) <= 1):
        continue
    corpus_result.append(c)
print(len(corpus_result))
```

```python
#카테고리마다 등장빈도 딕셔너리 생성
import re##
from collections import OrderedDict

def words_ranked(text):
    wordsRanked = {}
    for i in text:
        if i not in wordsRanked:
            wordsRanked[i] = 1
        else:
            wordsRanked[i] += 1
    wordsRanked = OrderedDict(sorted(wordsRanked.items(), key=lambda t: t[1], reverse=True))
    return wordsRanked
```

```python
#카테고리별 rank {단어 : 빈도}
books_rank = words_ranked(booksText)
self_rank = words_ranked(selfDev)
program_rank = words_ranked(program)
volunteer_rank = words_ranked(volunteer)
```

```python
#각 카테고리별로 빈도가 같으면?
#Cmax
source =""
target = -1
cmax = {}
for i in range(len(corpus_result)):
    temp_list =[]
    source = corpus_result[i]

    if source in books_rank:
        temp_list.append([books_rank[source], 'booksText'])
    if source in self_rank:
        temp_list.append([self_rank[source], 'selfDev'])
    if source in program_rank:
        temp_list.append([program_rank[source], 'program'])
    if source in volunteer_rank:
        temp_list.append([volunteer_rank[source], 'volunteer'])

    print(temp_list)
    index_list = []
    for j in range(len(temp_list)):
        index_list.append(temp_list[j][0])
    cmax[source] = {temp_list[index_list.index(max(index_list))][0], temp_list[index_list.index(max(index_list))][1]}
```

```python
#CTF
def get_ctf(word):
    res = -1
    temp = tuple(cmax[word])
    if(str(type(temp[0])) == "<class 'int'>"):
        res = temp[0]
    else:
        res = temp[1]
    return res
```

```python
#TF
def get_tf(doc, word):
    return doc.count(word)
```

```python
#IECDF
#Cothers 에 w가 없으면?
def get_iecdf(word):
    books_others = 298
    volunteer_others = 326
    program_others = 324
    self_others = 132

    cb = 0
    cp = 0
    cv = 0
    cs = 0

    others_count = -1
    exist_count = -1

    sour_category = ""
    temp = tuple(cmax[word])
    if(str(type(temp[0])) == "<class 'str'>"):
        sour_category = temp[0]
    else:
        sour_category = temp[1]

    for idd in books_doc:
        if word in idd:
            cb += 1
    for idd in volunteer_doc:
        if word in idd:
            cv += 1
    for idd in program_doc:
        if word in idd:
            cp += 1
    for idd in self_doc:
        if word in idd:
            cs += 1

    if(sour_category == 'booksText'):
        others_count = books_others
        exist_count = cp + cv + cs
    elif(sour_category == 'program'):
        others_count = program_others
        exist_count = cb + cv + cs
    elif(sour_category == 'selfDev'):
        others_count = self_others
        exist_count = cb + cp + cv
    elif(sour_category == 'volunteer'):
        others_count = volunteer_others
        exist_count = cb + cp + cs

    if(exist_count == 0):
        exist_count = 1

    return log(others_count/exist_count)
```

```python
#CDF
def get_cdf(word):
    word_count = -1
    doc_count = -1
    sour_category = ""
    temp = tuple(cmax[word])
    if(str(type(temp[0])) == "<class 'str'>"):
        sour_category = temp[0]
    else:
        sour_category = temp[1]

    cb = 0
    cp = 0
    cv = 0
    cs = 0

    if(sour_category == 'booksText'):
        for idd in books_doc:
            if word in idd:
                cb += 1
        word_count = cb
        doc_count = 62

    elif(sour_category == 'program'):
        for idd in program_doc:
            if word in idd:
                cp += 1
        word_count = cp
        doc_count = 36

    elif(sour_category == 'selfDev'):
        for idd in self_doc:
            if word in idd:
                cs += 1
        word_count = cs
        doc_count = 34

    elif(sour_category == 'volunteer'):
        for idd in volunteer_doc:
            if word in idd:
                cv += 1
        word_count = cv
        doc_count = 228


    return word_count/doc_count
```

```python
# IDF
def get_idf(word):
    N = len(merged_doc)
    df = 0
    for doc in merged_doc:
        if word in doc:
            df += 1
    return log(N/(df))
```

```python
#전체 단어 IDF 형태
idf = []
for word in corpus_result:
    idf.append(get_idf(word))
idf_ = pd.DataFrame(idf, index = corpus_result, columns = ["IDF"])
print(idf_)
```

```python
#전체 단어 cdf 형태
cdf = []
for word in corpus_result:
    cdf.append(get_cdf(word))
cdf_ = pd.DataFrame(cdf, index = corpus_result, columns =["CDF"])
print(cdf_)
```

```python
#전체 단어 iecdf 형태
iecdf = []
for word in corpus_result:
    iecdf.append(get_iecdf(word))
iecdf_ = pd.DataFrame(iecdf, index = corpus_result, columns =["IECDF"])
print(iecdf_)
```

```python
#전체 문서 TF_CTF_IECDF 행렬 구하기
import math
ctf_iecdf = []
for i in range(len(merged_doc)):
    ctf_iecdf.append([])
    d = merged_doc[i]
    for j in range(len(corpus_result)):
        t = corpus_result[j]
        tt = get_tf(d, t) * get_ctf(t) * get_iecdf(t)
        ctf_iecdf[-1].append(math.sqrt(tt))
    
```

```python
#TF_CTF_IECDF 행렬 확인
ctf_iecdf_ = pd.DataFrame(ctf_iecdf, columns = corpus_result)
print(ctf_iecdf_)
```

```python
#전체 문서 TF_CDF_IECDF 행렬 구하기
cdf_iecdf =  []
for i in range(len(merged_doc)):
    cdf_iecdf.append([])
    d = merged_doc[i]
    for j in range(len(corpus_result)):
        t = corpus_result[j]   
        cdf_iecdf[-1].append(get_tf(d, t) * get_cdf(t) * get_iecdf(t))
```

```python
#TF_CDF_IECDF 행렬 확인
cdf_iecdf_ = pd.DataFrame(cdf_iecdf, columns = corpus_result)
print(cdf_iecdf_)
```

```python
#전체 문서 TF_IDF 행렬 구하기
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfv = TfidfVectorizer().fit(corpus_result)
tf_idf_ = tfidfv.transform(corpus_result).toarray()
print(tf_idf_)
```

```python
#전체 문서 TF_CDF_IDF 행렬 구하기
cdf_idf = []
for i in range(len(merged_doc)):
    cdf_idf.append([])
    d = merged_doc[i]
    for j in range(len(corpus_result)):
        t = corpus_result[j]
        cdf_idf[-1].append(get_tf(d, t) * get_cdf(t) * get_idf(t))
```

```python
cdf_idf_ = pd.DataFrame(cdf_idf, columns = corpus_result)
print(cdf_idf_)
```
